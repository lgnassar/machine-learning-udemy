```python
# Import matplotlib and setup the figures to display within the notebook
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist();
```

```python
heart_disease["age"].plot.hist(bins=50);
```

```python
heart_disease.plot.hist(subplots=True);
```

```python
heart_disease.plot.hist(figsize=(10,30), subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])
ax.set(title='Heart Disease and Cholesterol Levels',
      xlabel='Age',
      ylabel='Cholesterol')
ax.legend(*scatter.legend_elements(),
          title='Target');
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

# ax0

scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
ax0.set(title='Heart Disease and Cholesterol Levels',
      xlabel='Age',
      ylabel='Cholesterol')
ax0.legend(*scatter.legend_elements(),
          title='Target')
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

# ax1

scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])
ax1.set(title='Heart Disease and Cholesterol Levels',
      xlabel='Age',
      ylabel='Max Heart Rate')
ax1.legend(*scatter.legend_elements(),
          title='Target')
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");

# main title

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
